```python
from transformers import BertTokenizer, BertForSequenceClassification
from transformers import pipeline
from transformers import TextClassificationPipeline, AutoModelForSequenceClassification, AutoTokenizer
```

```python
model_name = "ElKulako/cryptobert"
tokenizer = AutoTokenizer.from_pretrained(model_name, use_fast=True)
```

```python
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels = 3)
```

```python
import torch
```

```python
# Define batch size and initialize empty list to store predictions
batch_size = 32
all_predictions = []

model = model.to('cuda')

# Define a function to process a batch of texts
def process_batch(batch):
    max_length=256
    inputs = tokenizer(batch, padding=True, truncation=True, max_length=max_length, return_tensors='pt')

    # Move all tensors within the dictionary to GPU
    inputs = {key: value.to('cuda') for key, value in inputs.items()}

    with torch.no_grad():
        outputs = model(**inputs)

    return torch.argmax(outputs.logits, dim=1).cpu().tolist()
```

```python
# Process data in batches
for i in range(0, len(merged_df), batch_size):
    batch = merged_df['text'][i:i+batch_size].tolist()
    predictions = process_batch(batch)
    all_predictions.extend(predictions)
```

```python
merged_df['sentiments'] = all_predictions

# Visualize or save results
merged_df.shape
```
